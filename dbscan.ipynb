### DBSCAN CAN CLUSTER NON LINEAR DATA AND GOOD WITH OUTLIERS

```python
from sklearn.cluster import DBSCAN#importing dbscan
```

```python
from sklearn.datasets import make_moons
```

```python
import matplotlib.pyplot as plt
```

```python
X,y=make_moons(n_samples=100,noise=0.05)#adding noise too because dbscan can handle outliers
```

```python
plt.scatter(X[:,0],X[:,1])#as we can see a non linear datasets its like cut circle
```

```python
from sklearn.preprocessing import StandardScaler#scaling module
scaler= StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
dbscan=DBSCAN(eps=0.4)#taking radius 0.4 
```

```python
dbscan.fit(X_scaled)#fitting to x scaled
```

```python
dbscan.labels_#labels of all datapoint to 2 clusters as 0 and 1
```

```python
plt.scatter(X[:,0],X[:,1],c=dbscan.labels_)#using label to show cluster
```

```python
plt.scatter(X[:,0],X[:,1],c=y)#cluster with output
```
